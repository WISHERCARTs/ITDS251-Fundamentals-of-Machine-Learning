### 6. List and Dict

**List**

List เป็น data structure พื้นฐานของภาษา Python เปรียบได้กับเป็น Array หนึ่งมิดิ


```
# Creating a list
colors = ['red', 'green', 'blue']

# List can have mixed types, but NOT recommended
no_good = ['red', 1.1, 0]

# Using range()
print(list(range(10)))
print(list(range(3, 10)))

# Accessing a list
print(colors[0])  
# index เริ่มต้นที่ 0
# Accessing many element in a list
print(colors[1:3])

# Concatenating lists
print(colors + no_good)
```
**Dict**

Dictionary จะทำงานเหมือนกับ list เพียงแต่ว่า index ของเราจะไม่ใช่ตัวเลขที่เรียงลำดับ index สามารถเป็นค่า string หรือว่า ตัวอื่นๆที่ไม่เรียงลำดับได้ เราจะเรียกมันว่า `key` ส่วนค่าที่อยู่ประจำแต่ละ `key` เรียกว่า `value`

In [ ]:
```
# Creating a dict
myDict = {'peanut butter': 'jelly'}

# Accessing
print(myDict['peanut butter'])

# Creating an empty dict and assign key: value
myDict = {}
myDict['ham'] = 'bacon'

### 9. Function
เราสามารถสร้าง functions สำหรับ code ที่จะใช้งานประจำเพื่อลดการ copy-paste หลายๆรอบ ความจริงแล้ว print() หรือว่า range() ก็คือ function อันนึงที่มีมาให้แล้วใน python (built-in functions)

**Function Call**


```
# Calling a function:

function_name(argument1, argument2):
  print('this is an argument')
  print(argument1, argument2)

# keyword arguments
function_name('a', 'b')`
```
**Function Definition**


```
def foobar(arg1, arg2, optional_arg1=2):
    print(arg1)
    print(arg2)
    print(optional_arg1)
    return 42

__ = foobar(1, 2)
__ = foobar(arg2=10, arg1=2, optional_arg1=3)
```
**Functions from other packages**

หาเรามี functions (หรือองค์ประกอบอื่นๆ) จาก packages อื่น เราสามารถใช้คำสั่ง `import` เพื่อนำเอา code ใน package นั้นๆมาใช้งานได้ (เรียกว่า `module`)

```
