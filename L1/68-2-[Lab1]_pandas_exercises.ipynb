Let's check out what a version of our car_sales DataFrame might look like with missing values.

```python
car_sales_missing = pd.read_csv("https://raw.githubusercontent.com/TipGreenTea/ITDS251x2-ML/main/car-sales-missing-data%20.csv")
car_sales_missing
```

Notice the missing values are represented as `NaN` in pandas DataFrames.

Let's try fill them.

```python
# Fill the Odometer column missing values with the mean of the column inplace
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),
                                     inplace=True)
```
