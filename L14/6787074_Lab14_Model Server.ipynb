# REST API by flask (Server)
This notebook is a simple example of how to create a REST API using Flask. 

````python
# Import the Flask class from the flask module
from flask import Flask, request, jsonify
import pickle as pk
import numpy as np
````

## Load model

````python
# ---------------------------------
# YOUR CODE HERE
with open('scaler.pk', 'rb') as f:
    scaler = pk.load(f)
with open('model.pk', 'rb') as f:
    model = pk.load(f)
# ---------------------------------
````

## Initialize the Flask app
Then, we need to initialize the Flask app and create a route that will call the function we created.

````python
# Create a new instance of the Flask class
app = Flask(__name__)

# Define a route for the default URL and its behavior
@app.route('/', methods=['POST'])
def predict():
    
    # ----------------------------------
    # YOUR CODE HERE
    # 1. รับข้อมูล JSON ที่ส่งมาจาก Client
    data = request.get_json() 
    
    # 2. ดึงค่าจาก JSON ตามชื่อคอลัมน์ที่เราเลือกไว้ 4 ตัว (cp, thalachh, exng, oldpeak)
    # หมายเหตุ: ชื่อใน [''] ต้องตรงกับที่ Client จะส่งมานะครับ
    v1 = data['cp']
    v2 = data['thalachh']
    v3 = data['exng']
    v4 = data['oldpeak']
    
    # 3. เตรียมข้อมูลให้อยู่ในรูปแบบที่โมเดลต้องการ (เป็นลิสต์ซ้อนลิสต์หรือ 2D Array)
    x_input = np.array([[v1, v2, v3, v4]])
    
    # 4. ทำ Scaling (ต้องใช้ตัว scaler เดิมจากขั้นตอนที่แล้ว)
    x_input_scaled = scaler.transform(x_input)
    
    # 5. สั่งให้โมเดลทำนายผล
    prediction = model.predict(x_input_scaled)
    
    # 6. ส่งผลลัพธ์กลับไปหา Client เป็น JSON (แปลงผลลัพธ์ให้เป็น int ปกติก่อนส่ง)
    return jsonify({'result': int(prediction[0])})
    # ----------------------------------
    # ----------------------------------
````

## Run the app
Finally, we need to run the app.

````python
# Define a route for the default URL and its behavior
# Use port 5000 for the Flask app
# Set debug to False for production use

# ------------------------------------
# YOUR CODE HERE
if __name__ == '__main__':
    # เปิดเซิร์ฟเวอร์ที่ port 5000
    app.run(host='0.0.0.0', port=5000, debug=False)
# ------------------------------------
````
