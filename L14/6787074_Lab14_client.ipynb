# REST API by flask (Client)
This notebook is a simple example of how to use REST API by flask.

```python
# Import the necessary libraries
import requests, json, sys
```

## Create a post request
Create a post request to send data to the server. The data is a dictionary with the important information about the user. The server will return a response with the status code and the data that was sent.

```python
# Define the POST request function
# This function takes in the features values and the URL of the model server
# And returns the prediction result
def post_request(cp, thalachh, exng, oldpeak, url):
    # ----------------------------------
    # YOUR CODE HERE
    # 1. จัดเตรียมข้อมูลเป็น Dictionary ให้ตรงกับที่ Server รอรับ
    payload = {
        'cp': cp,
        'thalachh': thalachh,
        'exng': exng,
        'oldpeak': oldpeak
    }
    
    # 2. ส่งข้อมูลแบบ POST ไปที่ URL ของ Server
    response = requests.post(url, json=payload)
    
    # 3. รับคำตอบที่ได้จาก Server (ซึ่งเป็นก้อน JSON)
    return response.json()
    # ----------------------------------
```

# Try it yourself
You can try to run the following code in your local machine.

```python
# แก้ไข URL ให้ตรงกับเครื่องของคุณ
url = 'http://127.0.0.1:5000/' 
cp_val = 2
thalachh_val = 150
exng_val = 0
oldpeak_val = 1.5


# Call the function and print the result
result = post_request(cp_val, thalachh_val, exng_val, oldpeak_val, url)
print("คำทำนายจาก Server:", result)
```
