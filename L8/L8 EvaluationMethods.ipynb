# Evaluation Methods by Sklearn

This notebook demonstrates evaluation metrics for regression, classification, and clustering using Scikit-learn. Each section includes example data generation and metric calculations.

## Regression Evaluation
In this section, we calculate the following metrics for regression problems:
- Mean Squared Error (MSE)
- Mean Absolute Error (MAE)
- Root Mean Squared Error (RMSE)
- R² Score (Coefficient of Determination)

```python
# --- Regression Evaluation ---
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Generate 10 example data
y_true_reg = np.array([1, 2, 1, 3, 4, 5, 6, 8, 7, 9])
y_pred_reg = np.array([1, 1, 2, 3, 3, 6, 5, 7, 8, 9])

# Calculate metrics
mse = mean_squared_error(y_true_reg, y_pred_reg)
mae = mean_absolute_error(y_true_reg, y_pred_reg)
rmse = np.sqrt(mse)
r2 = r2_score(y_true_reg, y_pred_reg)

# Display results
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
print("R² Score:", r2)
```

## Classification Evaluation
In this section, we calculate the following metrics for classification problems:
- F1 Score
- Accuracy
- Precision
- Recall

```python
# --- Classification Evaluation ---
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score

# Generate 10 example data for 3 classes
y_true_clf = np.array([0, 1, 1, 2, 2, 0, 1, 2, 0, 1])
y_pred_clf = np.array([0, 1, 2, 2, 1, 0, 1, 2, 0, 1])

# Calculate metrics for Macro Average
fscore_M = f1_score(y_true_clf, y_pred_clf, average='macro')
precision_M = precision_score(y_true_clf, y_pred_clf, average='macro')
recall_M = recall_score(y_true_clf, y_pred_clf, average='macro')
# Display results
print("Macro Average")
print("F1 Score:", fscore_M)
print("Precision:", precision_M)
print("Recall:", recall_M)
print()

# Calculate metrics for Weighted Average
fscore_w = f1_score(y_true_clf, y_pred_clf, average='weighted')
precision_w = precision_score(y_true_clf, y_pred_clf, average='weighted')
recall_w = recall_score(y_true_clf, y_pred_clf, average='weighted')
# Display results
print("Weighted Average")
print("F1 Score:", fscore_w)
print("Precision:", precision_w)
print("Recall:", recall_w)
print()

# Calculate accuracy
accuracy = accuracy_score(y_true_clf, y_pred_clf)
print("Accuracy:", accuracy)
```

Confusion matrix is used to visualize the classification results.

```python
from sklearn.metrics import confusion_matrix

# confusion matrix
conf_matrix = confusion_matrix(y_true_clf, y_pred_clf)

# Display results
print("Confusion Matrix:")  
print(conf_matrix)
```

Report is generated using classification_report function.

```python
from sklearn.metrics import classification_report

# classification report
class_report = classification_report(y_true_clf, y_pred_clf)

# Display results
print("Classification Report:")
print(class_report)
```

## Clustering Evaluation
In this section, we calculate the following metrics for clustering problems:
- Silhouette Score
- Calinski-Harabasz Index
- Davies-Bouldin Index

Additionally, pairwise distances (Cosine Similarity and Euclidean Distance) are computed.

```python
# --- Clustering Evaluation ---
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from scipy.spatial.distance import cosine, euclidean
import matplotlib.pyplot as plt

# Generate example data
X_clust = np.array([[1, 2], [1, 3], [2, 3], [8, 7], [8, 8], [9, 8], [1, 2], [2, 3], [1, 3], [8, 7]])
labels_clust = np.array([0, 0, 0, 1, 1, 1, 0, 0, 0, 1])

# Plot data
plt.scatter(X_clust[:, 0], X_clust[:, 1], c=labels_clust, cmap='rainbow')
plt.show()
```

```python
# Calculate metrics
silhouette = silhouette_score(X_clust, labels_clust)
calinski_harabasz = calinski_harabasz_score(X_clust, labels_clust)
davies_bouldin = davies_bouldin_score(X_clust, labels_clust)

# Pairwise distances
cosine_similarities = [1 - cosine(X_clust[i], X_clust[j]) for i in range(len(X_clust)) for j in range(i+1, len(X_clust))]
euclidean_distances = [euclidean(X_clust[i], X_clust[j]) for i in range(len(X_clust)) for j in range(i+1, len(X_clust))]

# Display results
print("Silhouette Score:", silhouette)
print("Calinski-Harabasz Index:", calinski_harabasz)
print("Davies-Bouldin Index:", davies_bouldin)
print("Average Cosine Similarity:", np.mean(cosine_similarities))
print("Average Euclidean Distance:", np.mean(euclidean_distances))
```
